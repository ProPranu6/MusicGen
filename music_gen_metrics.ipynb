```python
import matplotlib.pyplot as plt
from music21 import converter, note, stream, pitch

def plot_pitch_class_histogram(midi_file):
    # Load MIDI file
    score = converter.parse(midi_file)

    # Extract pitch classes
    pitch_classes = [0] * 12
    for element in score.recurse():
        if isinstance(element, note.Note):
            pc = element.pitch.pitchClass
            pitch_classes[pc] += 1

    # Plot pitch class histogram
    plt.figure(figsize=(8, 6))
    plt.bar(range(12), pitch_classes, color='skyblue')
    plt.title('Pitch Class Histogram')
    plt.xlabel('Pitch Class')
    plt.ylabel('Frequency')
    plt.xticks(range(12), ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B'])
    plt.show()

def plot_music_sheet(midi_file):
    # Load MIDI file
    score = converter.parse(midi_file)

    # Plot music sheet
    score.show('lily.pdf')

# Example usage
midi_file = 'generations_set/generated_track_transformer_14.mid'

# Plot pitch class histogram
plot_pitch_class_histogram(midi_file)

# Plot music sheet
plot_music_sheet(midi_file)
```
